# Missing Values for categorial variable

#### As in the earlier sessions we have seen numeric variable  conatin a lot of outliers so we replace missing val with Median

```python
## Let us capture all the nan values 
## First lets handle Categorical features which are missing
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().mean()>0 and dataset[feature].dtypes=="O"]
for feature in feature_nan:
     print("{}: {}% missing values".format(feature,round(dataset[feature].isnull().mean()*100,4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna("Missing")
    return data
dataset=replace_cat_feature(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

# Missing Values for Numeric variable

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().mean()>0 and dataset[feature].dtype != "O"]
for feature in numerical_with_nan:
     print("{}: {}% missing values".format(feature,round(dataset[feature].isnull().mean()*100,4)))
```

```python
## Replacing the numerical Missing Values
for feature in numerical_with_nan:
    # We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
     ## create a new feature to capture nan values
    dataset[feature+"NaN"]=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)

dataset[numerical_with_nan].isnull().sum()
```

# Handling Rare Categorical Feature
We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype == "O"]
categorical_features
```

## For Reference not any actual role in the project

```python
dataset.groupby("HeatingQC")["SalePrice"].count()
```

## calculating the catorical variables that are present less than 1% of the observations

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    temp=temp[temp>.01].index
    # Below line mean if  dataset[feature] is in temp then  dataset[feature]= dataset[feature] else not then  dataset[feature] 
    # is replace by Rare_Var
    dataset[feature]=np.where(dataset[feature].isin(temp),dataset[feature],"Rare_var")
```

```python
dataset.head(31)
```

 # Converting categorial variable to numeric variable

```python
dataset.groupby([feature])['SalePrice'].mean().sort_values().index
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

# Feature Scaling

```python
# As we dont require ID and Sales Price as id is ust an incrimenatl value and saleprice is to be predicted variable
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

## Note
    * Checked with both Min max and standardizaton as there are less num of negative values in the data we apply Min_Max_scaling but this is not the case that we always go with standariztion if there are more negative values, if this Kish applied both but Min_Max scaling give high accuracy

```python
scaler.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
