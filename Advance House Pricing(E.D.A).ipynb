## Missing Values

```python
def miss_val(dataset):
    val=0
    for i in dataset.columns:
        if dataset[i].isnull().mean()>0:
            val=np.round(dataset[i].isnull().mean(),4)
            print(i,"Missing Val is:-- ",val)
miss_val(dataset)
```

#### Since they are many missing values, we need to find the relationship between missing values and Sales Price
Let's plot some diagram for this relationship

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
    data = dataset.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    # 1 for Nan and 0 for non-missing
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    
    # let's calculate the median SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()

    # Note 1 nd 0 is applied to data locally only not outsde the scope of thi for loop
```

* Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section.
* Here 1 mean missing as you see in some cases there is high graphs for missing values with salesPrice as in "LotFrontage" varibale and amany more

## For Continous  features


```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature +year_feature+ ["Id"]]
print("The continuous_feature are: {}".format(len(continuous_feature)))
```

```python
data[continuous_feature].head()
```

### In contionous variable we find the distribution of the variable

```python
## Lets analyse the continuous values by creating histograms to understand the distribution
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

## Notice in here most of the variable have Skeweness

### To treat this we use Log transformation

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data["SalePrice"]=np.log(data["SalePrice"])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

## Observation
    * Here you see after applying log transformation the variable folow a liner relation with the salesprce
